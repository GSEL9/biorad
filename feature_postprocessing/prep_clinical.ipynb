```python
clinical.index.name = 'patient'
clinical.columns = [
    'Age', 'Sex', 
    'ICD-10', 
    'T Stage', 'N Stage', 'Stage', 
    'Histology', 'HPV', 
    'ECOG', 'Charlson', 
    'Years Smoking', 
    'Naxogin Days', 'Cisplatin'
]
clinical.head()
```

```python
# NOTE: Stage feature req extra attention.
to_encode = [
    'Sex', 
    'ICD-10', 
    'T Stage', 
    'N Stage', 
    'Histology', 
    'HPV', 
    'ECOG', 
    #'Charlson', 
    'Cisplatin'
]

for num, var_to_encode in enumerate(to_encode):
    dummy_var = pd.get_dummies(clinical[var_to_encode], drop_first=True)
    dummy_var.columns = [
        '{}_{}'.format(var_to_encode, column) for column in dummy_var.columns
    ]
    clinical[dummy_var.columns] = dummy_var
```

```python
clinical.columns
```

```python
clinical.head()
```

```python
# Drop already encdoded features
clinical.drop(to_encode, axis=1, inplace=True)
clinical.head()
```

```python
clinical.Stage.replace(['4a', '4b', '4c', '4a?'], 4, inplace=True)
stadium_enc = pd.get_dummies(clinical.loc[:, 'Stage'], drop_first=True)
for col in stadium_enc:
    clinical['Stage {}'.format(col)] = stadium_enc[col]

clinical.head()
```

```python
clinical.drop('Stage', axis=1, inplace=True)
clinical.head()
```
