# PET Parameters

```python
import numpy as np
import pandas as pd
```

```python
true_index = np.load('./../../../data/source/patient_id.npy')
true_index[:5]
```

```python
raw_pet_params = pd.read_excel(
    './../../../data/source/raw_tabular/pet_params.xlsx', 
    index_col=0
)
raw_pet_params.head()
```

```python
raw_pet_params.shape
```

```python
pet_params = raw_pet_params.loc[true_index, :]
```

```python
pet_params.index = true_index
pet_params.index.name = 'patient'
pet_params.columns = [col[1:-1] for col in pet_params.columns]
pet_params.head()
```

```python
pet_params.drop(pet_params.columns[-2:], axis=1, inplace=True)
pet_params.head()
```

```python
# Removing features extracted with PyRadiomics.
pet_params.drop(['SUVmax', 'volume'], axis=1, inplace=True)
pet_params.head()
```

```python
pet_params.to_csv('./../../../data/to_analysis/pet_params.csv')
```
