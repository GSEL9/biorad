```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.utils.fixes import signature
from sklearn.metrics import precision_recall_curve

sns.set()
%matplotlib inline
```

```python
def plot_precision_recall(precisions, recalls):
    
    avg_precision = np.mean(precisions, axis=0)
    avg_recall = np.mean(recalls, axis=0)
    
    step_kwargs = (
        {'step': 'post'} if 'step' in signature(plt.fill_between).parameters else {}
    )
    plt.step(
        avg_recall, avg_precision, color='b', alpha=0.2, where='post'
    )
    plt.fill_between(avg_recall, avg_precision, alpha=0.2, color='b', **step_kwargs)

    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 0.8])
    plt.xlim([0.0, 1.0])
```

# Removed Broken Slices: Clinical Only

```python
precisions = np.load(
    './../../data_source/results/removed_broken_slices/precision_recall_curve/clinical_only/precisions.npy'
)
recalls = np.load(
    './../../data_source/results/removed_broken_slices/precision_recall_curve/clinical_only/recalls.npy'
)
plt.figure(figsize=(10, 10))
plot_precision_recall(precisions, recalls)
```
